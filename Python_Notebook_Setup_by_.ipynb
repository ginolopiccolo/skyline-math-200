Research Question # __ : Is  there association between drinking a beverage with caffeine in the last 24 hours and having trouble falling asleep

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import researchpy as rp
```

```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head()
```

Codebook 1 Screen Shot 2023-05-01 at 12.27.55 PM.png

```python
#Manage data for variable 1
df ['H4GH12'].value_counts()
```

```python
df['CAF_BEV']=df['H4GH12']
df['CAF_BEV'].value_counts()
```

```python
df['CAF_BEV'] = df ['CAF_BEV'].loc[df['CAF_BEV']!=' ']
df['CAF_BEV'] = df ['CAF_BEV'].loc[df['CAF_BEV']!='8']
df['CAF_BEV'].value_counts()
```

```python
df['CAF_BEV'].replace(['0','1'], ['NO', 'YES',] ,inplace = True)
df['CAF_BEV'].value_counts ()
```

```python
rp.summary_cat(df['CAF_BEV'])
```

```python
#Graphs and tables or statistics
plt.title("Have you had a caffeinated beverage in the last 24 hours ")
df['CAF_BEV'].value_counts().plot(kind ='pie', autopct='%1.1f%%', colors=['olivedrab', 'rosybrown', 'gray', 'saddlebrown','blue'])
```

```python
plt.title("Have you had a caffeinated beverage in the last 24 hours?")
df['CAF_BEV'].value_counts(normalize = True).plot(kind ='bar', color = 'green', width = 0.2)
```

```python
#Manage data for variable 1
df ['H4SP5'].value_counts()
```

```python
df['TRBL_SLEEP']=df['H4SP5']
df['TRBL_SLEEP'].value_counts()
```

```python
df['TRBL_SLEEP'] = df['TRBL_SLEEP'].loc[ df['TRBL_SLEEP'] !=' ' ]
df['TRBL_SLEEP'] = df['TRBL_SLEEP'].loc[ df['TRBL_SLEEP'] !='5' ]
df['TRBL_SLEEP'] = df['TRBL_SLEEP'].loc[ df['TRBL_SLEEP'] !='8' ]
df['TRBL_SLEEP'].value_counts()
```

```python
df['TRBL_SLEEP'].replace(['0','1','2','3','4',], ['Never in the past 4 weeks', 'less than one week','1 or 2 times a week','3 or 4 times a week','5 or more times a week'] ,inplace = True)
df['TRBL_SLEEP'].value_counts ()
```

```python
rp.summary_cat(df['TRBL_SLEEP'])
```

```python
#Graphs and tables or statistics
plt.title("How often did you have trouble falling asleep? ")
df['TRBL_SLEEP'].value_counts().plot(kind ='pie', autopct='%1.1f%%', colors=['olivedrab', 'rosybrown', 'gray', 'saddlebrown','blue'])
```

```python
plt.ylabel("Percent")
plt.title("How often do you find trouble falling asleep?")
df['TRBL_SLEEP'].value_counts(normalize = True).plot(kind ='bar', color = 'blue', width = 0.2, alpha = 0.1)
```
